```python
import pandas as pd
import numpy as np 
```

```python
raw_gdp = pd.read_csv(r'D:\Canadian Business Analysis Project\data\3610044901-eng.csv', skiprows=11)
```

```python
raw_gdp = raw_gdp.replace("..", np.nan)
```

```python
raw_gdp = raw_gdp.dropna(axis=1, how='all')
```

```python
raw_gdp.head()
```

```python
raw_gdp.tail(15)
```

```python
naics_col = "North American Industry Classification System (NAICS) 3"
quarter_cols = [c for c in raw_gdp.columns if str(c).strip().startswith("Q")] # Get all quarter columns
```

```python
raw_gdp_clean = raw_gdp.dropna(subset=quarter_cols, how="all").copy() # Drop rows where all quarter columns are NaN
```

```python
raw_gdp_clean.tail(15)
```

```python
gdp_num = raw_gdp.copy()

for c in quarter_cols:
    gdp_num[c] = (
        gdp_num[c]
        .astype(str)
        .str.replace(",", "", regex=False)  # remove thousands separator from numbers under Quarter columns
    )
    gdp_num[c] = pd.to_numeric(gdp_num[c], errors="coerce") 
```

```python
min_numeric_quarters = 2
mask = gdp_num[quarter_cols].notna().sum(axis=1) >= min_numeric_quarters # At least 2 numeric quarter values
raw_gdp_clean = raw_gdp.loc[mask].copy() # Keep only rows that meet the criteria
```

```python
raw_gdp_clean.tail(5)
```

```python
raw_gdp_clean.info()
```

```python
raw_gdp_clean.describe()
```

```python
gdp_long = raw_gdp_clean.melt(   # Convert from wide to long format
    id_vars=[naics_col], # Keep NAICS column as identifier
    var_name="quarter",  # Name for the new 'variable' column
    value_name="gdp" # Name for the new 'value' column
)
```

```python
gdp_long
```

```python
gdp_long["year"] = gdp_long["quarter"].str[-4:].astype(int)
```

```python
gdp_long["quarter_num"] = (
    gdp_long["quarter"]
    .str.extract(r"Q([1-4])") # Extract quarter number using regex
    .astype(int)
)
```

```python
gdp_long = gdp_long.drop(columns=["quarter"])
```

```python
gdp_long.head()
```

```python
gdp_long = gdp_long.rename(columns={naics_col: "industry"})
```

```python
gdp_long.head()
```

```python
gdp_long = gdp_long.sort_values(by=["industry", "year", "quarter_num"]).reset_index(drop=True)
```

```python
gdp_long["gdp"] = (
    gdp_long["gdp"]
    .astype(str)
    .str.replace(",", "", regex=False)
)
```

```python
gdp_long["gdp"] = pd.to_numeric(gdp_long["gdp"], errors="coerce") # Ensure 'gdp' column is numeric
```

```python
gdp_long["gdp_qoq_growth"] = gdp_long.groupby("industry")["gdp"].pct_change() # Calculate quarter-over-quarter GDP growth per industry
```

```python
gdp_long.head(10)
```

```python
gdp_long["gdp_volatility_4q"] = (
    gdp_long
    .groupby("industry")["gdp_qoq_growth"]
    .rolling(4) # 4-quarter rolling standard deviation
    .std() # Calculate standard deviation
    .reset_index(level=0, drop=True) # Drop the grouping index
)
```

```python
gdp_long.head(15)
```

```python
gdp_long.to_csv(r'D:\Canadian Business Analysis Project\data\processed_gdp_data.csv', index=False)
```

```python
cpi_raw = pd.read_csv(r'D:\Canadian Business Analysis Project\data\1810000401-eng.csv', skiprows=9)
```

```python
cpi_raw.head()
```

```python
cpi_raw = cpi_raw.replace("..", np.nan)
```

```python
cpi_raw = cpi_raw.dropna(axis=1, how='all')
```

```python
cpi_raw = cpi_raw.drop(0, axis=0)
```

```python
product_col = "Products and product groups 3 4"  

cpi_all = cpi_raw[cpi_raw[product_col].str.contains("All-items", case=False)].copy() # Keep only 'All-items' rows
```

```python
cpi_all = cpi_all.drop([11, 12], axis=0)
```

```python
cpi_all
```

```python
import re

month_cols = [
    c for c in cpi_all.columns
    if isinstance(c, str) and re.match(r"^[A-Za-z]+ \d{4}$", c) # re.match matches at the start of the string
]
```

```python
cpi_long = cpi_all.melt(
    id_vars=[product_col],
    value_vars=month_cols,
    var_name="date",
    value_name="cpi"
)
```

```python
cpi_long = cpi_long.drop(columns=["Products and product groups 3 4"])
```

```python
cpi_long.head()
```

```python
cpi_long["date"] = pd.to_datetime(
    cpi_long["date"],
    format="%B %Y" # Month Year format
)
```

```python
cpi_long["year"] = cpi_long["date"].dt.year
cpi_long["quarter_num"] = cpi_long["date"].dt.to_period("Q").dt.quarter
```

```python
cpi_long.head()
```

```python
cpi_long["cpi"] = (
    cpi_long["cpi"]
    .astype(str)
    .str.strip() # Remove leading/trailing whitespace
)

cpi_long["cpi"] = pd.to_numeric(
    cpi_long["cpi"],
    errors="coerce"
)
```

```python
cpi_q = (
    cpi_long
    .groupby(["year", "quarter_num"], as_index=False)
    .agg(cpi=("cpi", "mean")) # Average CPI per quarter
)
```

```python
cpi_q.head()
```

```python
cpi_q = cpi_q.sort_values(["year", "quarter_num"]).reset_index(drop=True)

cpi_q["cpi_qoq_growth"] = cpi_q["cpi"].pct_change()
cpi_q["cpi_yoy_growth"] = cpi_q["cpi"].pct_change(4) # Year-over-year growth (4 quarters)
```

```python
cpi_q["cpi_volatility_4q"] = cpi_q["cpi_qoq_growth"].rolling(4).std() # 4-quarter rolling std dev
```

```python
cpi_q.head()
```

```python
cpi_q.to_csv(r'D:\Canadian Business Analysis Project\data\processed_cpi_data.csv', index=False)
```

```python
acf_data = pd.read_csv(r'D:\Canadian Business Analysis Project\data\ds_accomfs_sect.csv', skiprows=11)
```

```python
acf_data = acf_data.drop(0, axis=0)
```

```python
acf_data = acf_data.replace("..", np.nan)
acf_data = acf_data.dropna(axis=1, how='all')
```

```python
acf_data.tail()
```

```python
acf_data.head(10)
```

```python
month_cols = acf_data.columns[1:]
month_cols
```

```python
acf_num = acf_data.copy()

for c in month_cols:
    acf_num[c] = (
        acf_num[c]
        .astype(str)
        .str.replace("\u00a0", "", regex=False)  # non-breaking space
        .str.replace(",", "", regex=False)       # thousands separator
        .str.replace(" ", "", regex=False)       # normal spaces
        .replace({"..": None, "": None, "": None, "nan": None})
    )
    acf_num[c] = pd.to_numeric(acf_num[c], errors="coerce")
```

```python
acf_num
```

```python
MIN_MONTHS = 2

mask = acf_num[month_cols].notna().sum(axis=1) >= MIN_MONTHS
acf_clean = acf_data.loc[mask].copy()
```

```python
acf_clean["Business dynamics measure"].unique()
```

```python
acf_clean
```

```python
acf_closures = acf_clean[
    acf_clean["Business dynamics measure"]
      .str.lower()
      .str.contains("closing")
].copy()
```

```python
acf_closures
```

```python
id_col = "Business dynamics measure"

acf_long = acf_closures.melt(
    id_vars=[id_col],
    value_vars=month_cols,
    var_name="month",
    value_name="closures"
)
```

```python
acf_long["year"] = (
    acf_long["month"]
    .str.extract(r"(\d{4})")
    .astype(int)
)
```

```python
acf_long["closures"] = (
    acf_long["closures"]
      .astype(str)
      .str.replace("\u00a0", "", regex=False)  # non-breaking spaces
      .str.replace(",", "", regex=False)       # thousands separators
      .str.strip()
)

acf_long["closures"] = pd.to_numeric(acf_long["closures"], errors="coerce")
```

```python
acf_yearly = (
    acf_long
    .groupby("year", as_index=False)
    .agg(closures=("closures", "sum"))
)
acf_yearly
```

```python
acf_yearly = acf_yearly.sort_values("year")

acf_yearly["closures_lag1"] = acf_yearly["closures"].shift(1) # Lagged closures by 1 year since we dont use same year data to predict same year insolvencies
```

```python
acf_yearly
```

```python
acf_yearly = acf_yearly.dropna(subset=["closures_lag1"]) # Drop rows where lagged closures is NaN (i.e., the first year)
```

```python
acf_yearly
```

```python
acf_yearly.to_csv(r'D:\Canadian Business Analysis Project\data\processed_acf_closures_yearly.csv', index=False)
```
